#### Step 2: Search density coordenates without assigned schools

```python
filtro = data2.loc[data2['school_index'].isna()] # Coordenates of density without school assigned
filtro['lon'] = filtro['lon'].round(2)
filtro['lat'] = filtro['lat'].round(2)
filtro
```

```python
agrup = filtro.groupby(['lon','lat'], as_index=False).sum().reset_index(drop=True)
agrup = agrup.loc[agrup['densidad'] > 620]
agrup.sort_values(['densidad'], ascending=False)
```

```python
agrup['escuelas'] = round((agrup['densidad'] / 620))
agrup.sort_values(['densidad'], ascending=False)
```

Escuelas para puntos de densidad mayores a 620 personas

```python
lon_list = agrup.apply(lambda row: [row['lon']] * int(row['escuelas']), axis=1).tolist()
lat_list = agrup.apply(lambda row: [row['lat']] * int(row['escuelas']), axis=1).tolist()

# Obtener la longitud total de la lista replicada
total_length = sum(len(x) for x in lon_list)

# Crear un nuevo DataFrame a partir de las listas replicadas
new_df = pd.DataFrame({
    'lon': [item for sublist in lon_list for item in sublist],
    'lat': [item for sublist in lat_list for item in sublist]
})

new_df
```

```python
agrup2 = filtro.groupby(['lon','lat'], as_index=False).sum().reset_index(drop=True)
agrup2 = agrup2.loc[agrup2['densidad'] <= 620]
agrup2.sort_values(['densidad'], ascending=False)
```

```python
grouped_sum = 0
grouped_values = []
result = []

# Iterar sobre los valores de la columna 'densidad'
for value in agrup2['densidad']:
    if grouped_sum + value <= 620:
        grouped_sum += value
        grouped_values.append(value)
    else:
        result.append(sum(grouped_values))
        grouped_values = [value]
        grouped_sum = value

# Agregar el último grupo si hay valores restantes
if grouped_values:
    result.append(sum(grouped_values))

# Crear un DataFrame a partir de los resultados
result_df = pd.DataFrame({'Grouped_Sum': result})

# Mostrar el DataFrame resultante
print(result_df.sum())
```

```python
grouped_sum = 0
grouped_values = []
result = []

# Crear una copia del DataFrame original
result_df = agrup2.copy()

# Iterar sobre los valores de la columna 'densidad'
for index, value in enumerate(agrup2['densidad']):
    if grouped_sum + value <= 620:
        grouped_sum += value
        grouped_values.append(value)
    else:
        result_df.loc[index, 'Grouped_Sum'] = sum(grouped_values)
        grouped_values = [value]
        grouped_sum = value

# Agregar el último grupo si hay valores restantes
if grouped_values:
    result_df.loc[len(agrup2), 'Grouped_Sum'] = sum(grouped_values)

# Calcular el promedio de las otras columnas en el DataFrame resultante
result_mean = result_df.groupby('Grouped_Sum', as_index=False).mean()

# Fusionar los resultados con el DataFrame original en la columna de índice
result_df = pd.merge(result_df, result_mean, on='Grouped_Sum', how='left')

# Mostrar el DataFrame resultante
filtro_result_df = result_df.loc[result_df['Grouped_Sum'].notna()]
filtro_result_df
```

```python
old_schools = pd.read_excel('schoolsandtheircoordinates2020.xlsx')
sample = old_schools.sample(766)
```

```python
add_value = 0.05

# Calcular los nuevos valores para 'latitude' y 'longitude'
random_value_lat = random.uniform(-add_value, add_value)
random_value_lon = random.uniform(-add_value, add_value)

new_latitude = sample['latitude'] + random_value_lat
new_longitude = sample['longitude'] + random_value_lon

# Ajustar los valores si exceden los límites
min_lat, max_lat = old_schools['latitude'].min(), old_schools['latitude'].max()
min_lon, max_lon = old_schools['longitude'].min(), old_schools['longitude'].max()

sample['latitude'] = new_latitude.apply(lambda x: min(max(x, min_lat), max_lat))
sample['longitude'] = new_longitude.apply(lambda x: min(max(x, min_lon), max_lon))
```
